## Fare Distribution

```python
plt.figure(figsize = (4,3))
sns.histplot(x="fare", data=df, kde=True, color="", bins=20)
plt.title("Fare Distribution", fontsize=16, fontweight="bold")
plt.xlabel("Fare ($)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Passenger Count by Embarkation Town

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = df.embark_town)
plt.ax = sns.countplot(x="embark_town", data=df, hue ="who", palette="pastel")
plt.title("Passenger Count by Embarkation Town", fontsize=16, fontweight="bold")
plt.xlabel("Embarkation Town", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type="edge", fontsize=10)

plt.show()
```

## Passenger Class by Sex and Survival

```python
plt.figure(figsize = (4,3))
sns.violinplot(x="sex", y="pclass", hue="survived", data=df, split=True, inner="quartile", palette="muted")
plt.title("Violin Plot: Passenger Class by Sex and Survival", fontsize=16, fontweight="bold")
plt.xlabel("Sex", fontsize=12)
plt.ylabel("Passenger Class", fontsize=12)
plt.legend(title="Survived", title_fontsize=12, fontsize=10)
plt.show()
```

## "Passenger Count by Class and Who

```python
class_counts = df.groupby(['class', 'who']).size().unstack()
class_counts.plot(kind="bar", stacked=True, color=["#FF9999", "#66B2FF", "#99FF99"], figsize=(5, 5))
plt.title("Passenger Count by Class and Who", fontsize=16, fontweight="bold")
plt.xlabel("Class", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Who", title_fontsize=12, fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

## Comparison of adults males vs. Females Traveling Alone

```python
plt.figure(figsize = (4,4))
sns.barplot(x="adult_male", y="alone", data=df, hue="sex",palette="pastel")
plt.title('Comparison of Adult Males vs. Females Traveling Alone')
plt.xlabel('Adult Male (True/False)')
plt.ylabel('Alone (True/False)')
plt.legend(title="Gender", loc="upper left") 
plt.show()
```
